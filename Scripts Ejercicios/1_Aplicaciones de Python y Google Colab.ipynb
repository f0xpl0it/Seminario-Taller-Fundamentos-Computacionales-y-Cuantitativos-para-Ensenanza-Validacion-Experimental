## 1.4. Variables y tipos de datos

En Python, las variables se crean automáticamente al asignarles un valor. No necesitas declarar el tipo.

```python
x = 10       # entero
y = 3.14     # decimal (float)
nombre = "Ana"  # cadena de texto
activo = True   # booleano
```

Caso especial para números complesjos.

```python
# Definir un número complejo
z = 3 + 5j
print(z)
# Salida: (3+5j)

# Acceder a las partes real e imaginaria
print(z.real)
# Salida: 3.0
print(z.imag)
# Salida: 5.0
```

## 1.9. Bucles: `for` y `while`

**For:** se usa para repetir un bloque un número de veces.

```python
for i in range(5):
    print(i)
```

**While:** se repite mientras una condición sea verdadera.

```python
x = 0
while x < 5:
    print(x)
    x += 1
```

# 4. Análisis Estadístico y Numérico Aplicado

<a id="4.1"></a>
## 4.1. Matriz de Vandermonde

Es una matriz usada en álgebra y polinomios. Se usa comúnmente para ajustar modelos polinómicos o resolver sistemas de ecuaciones con polinomios.

```python
import numpy as np
x = np.array([1, 2, 3])
vander = np.vander(x, increasing=True)
print(vander)
```

<a id="4.2"></a>
## 4.2. Regresión Lineal Simple

Relaciona dos variables con una línea recta.

```python
from sklearn.linear_model import LinearRegression

X = np.array([[1], [2], [3]])
y = np.array([2, 4, 5])

model = LinearRegression()
model.fit(X, y)
print(model.coef_, model.intercept_)
```

<a id="4.3"></a>
## 4.3. ANOVA: Análisis de Varianza

Compara medias entre 3 o más grupos.

```python
import scipy.stats as stats

grupo1 = [4,5,6]
grupo2 = [5,6,7]
grupo3 = [8,9,10]

f, p = stats.f_oneway(grupo1, grupo2, grupo3)
print("p-value:", p)
```

<a id="4.4"></a>
## 4.4. Prueba de Tukey (post-hoc)

Se aplica luego de una ANOVA cuando hay diferencias significativas, para saber entre qué grupos existen esas diferencias.

```python
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd

df = pd.DataFrame({
    'valor': [5,6,7,5,6,7,8,9,10],
    'grupo': ['A']*3 + ['B']*3 + ['C']*3
})

tukey = pairwise_tukeyhsd(endog=df['valor'], groups=df['grupo'], alpha=0.05)
print(tukey)
```

<a id="4.5"></a>
## 4.5. PCA: Análisis de Componentes Principales

Es una técnica para reducir la cantidad de variables en un conjunto de datos, transformándolas en nuevas variables llamadas componentes principales, que conservan la mayor varianza posible.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X = np.array([[1,2], [3,4], [5,6]])
X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_scaled)
print(X_pca)
```

<a id="4.6"></a>
## 4.6. Modelo de Crecimiento Poblacional

Es un modelo exponencial que representa cómo una población crece con el tiempo según una tasa constante. Se aplica también en economía, epidemiología, etc.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 100)
P0 = 100
r = 0.3
P = P0 * np.exp(r * t)

plt.plot(t, P)
plt.title("Crecimiento poblacional")
plt.xlabel("Tiempo")
plt.ylabel("Población")
plt.show()
```

## 5.1. Diferencia entre `import` y `from ... import`

`import`

Carga todo el módulo, y para usar sus funciones debes anteponer el nombre del módulo.

```python
import math
print(math.sqrt(25))
```

`from ... import`

Importa solo una parte específica del módulo, lo que te permite usarla directamente.

```python
from math import sqrt
print(sqrt(25))
```
