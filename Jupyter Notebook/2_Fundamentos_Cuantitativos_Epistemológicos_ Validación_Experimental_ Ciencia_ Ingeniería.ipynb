# 1. Introducción

La epistemología es la rama de la filosofía que estudia el conocimiento: cómo se obtiene, valida y estructura.

 * **Definición:** La epistemología es el estudio de los fundamentos, la naturaleza y el alcance del conocimiento. En la investigación cuantitativa, se enfoca en cómo se construye el conocimiento a través de datos numéricos y cómo estos datos son procesados, analizados y validados.

* **Fundamento:** La investigación cuantitativa se basa en el paradigma positivista, que sostiene que la realidad es objetiva y puede ser medida de manera precisa a través de variables observables y manipulables.

<div class="alert alert-block alert-info">
La investigación cuantitativa se basa en el análisis de datos numéricos utilizando diversas herramientas estadísticas para establecer patrones, relaciones y conclusiones. Algunas de las pruebas estadísticas más utilizadas en este enfoque incluyen:

1. **F de Fisher:** Prueba estadística utilizada para comparar la variabilidad entre dos o más grupos y evaluar si las diferencias entre sus medias son significativas. Se aplica, por ejemplo, en el análisis de varianza (ANOVA).
2. **T de Student:** Utilizada para comparar las medias de dos grupos independientes y determinar si hay diferencias significativas entre ellos. Es especialmente útil cuando las muestras son pequeñas.
3. **Análisis de Varianza (ANOVA):** Técnica que compara las medias de tres o más grupos para determinar si al menos uno de ellos es significativamente diferente.
4. **Comparación Múltiple de Medias (Tukey):** Método post-hoc utilizado tras un ANOVA para identificar qué pares de grupos son significativamente diferentes entre sí.
5. **Regresión Lineal:** Herramienta estadística que modela la relación entre una variable dependiente y una o más variables independientes.
6. **Correlación de Pearson:** Mide la fuerza y dirección de la relación lineal entre dos variables.
7. **Análisis de Componentes Principales (PCA):** Técnica utilizada para reducir la dimensionalidad de datos multivariados, identificando las direcciones de mayor variabilidad.
8. **Intervalos de Confianza:** Rango en el que se estima que se encuentra un parámetro poblacional, proporcionando una medida de la precisión de las estimaciones.
9. **Prueba de Hipótesis:** Procedimientos estadísticos usados para evaluar afirmaciones sobre parámetros poblacionales.
10. **Métodos No Paramétricos:** Técnicas estadísticas útiles cuando los datos no siguen una distribución normal, como la prueba de Mann-Whitney U.

## **1. Prueba T de Student**
> La prueba t de Student se utiliza para comparar las medias de dos grupos independientes. Aquí te muestro

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Datos de ejemplo (dos grupos)
grupo_1 = [23, 21, 19, 24, 30, 25, 22, 24, 23, 27]
grupo_2 = [30, 29, 31, 35, 38, 40, 33, 30, 32, 35]

# Realizamos la prueba T de Student
t_stat, p_value = stats.ttest_ind(grupo_1, grupo_2)

# Imprimir resultados
print("Estadístico T:", t_stat)
print("Valor p:", p_value)

# Interpretación de los resultados
alpha = 0.05  # Nivel de significancia
if p_value < alpha:
    print("Hay una diferencia significativa entre los dos grupos.")
else:
    print("No hay una diferencia significativa entre los dos grupos.")

# Gráficos
plt.figure(figsize=(12, 6))

# Boxplot para visualizar las distribuciones de ambos grupos
plt.subplot(1, 2, 1)
plt.boxplot([grupo_1, grupo_2], labels=["Grupo 1", "Grupo 2"], patch_artist=True, boxprops=dict(facecolor="skyblue", color="blue"))
plt.title("Distribución de los Grupos")
plt.ylabel("Valor")
plt.grid(True)

# Gráfico de barras con las medias de los grupos
plt.subplot(1, 2, 2)
group_means = [np.mean(grupo_1), np.mean(grupo_2)]
plt.bar(["Grupo 1", "Grupo 2"], group_means, color=['blue', 'orange'])
plt.title("Media de los Grupos")
plt.ylabel("Media")
plt.grid(True)

plt.tight_layout()
plt.show()
````

* Comparamos dos grupos independientes ```(grupo_1 y grupo_2)```.
* El valor p nos indica si las medias de los dos grupos son significativamente diferentes.
* Si el valor p es menor que el nivel de significancia $(\alpha = 0.05)$, podemos rechazar la hipótesis nula de que las medias son iguales.
* **Boxplot:** Muestra la distribución de los datos, incluyendo la mediana, el rango intercuartílico y los posibles valores atípicos. Esto te ayudará a comparar la dispersión de ambos grupos.
* **Gráfico de Barras:** Muestra las medias de los dos grupos, lo que te permite comparar visualmente las diferencias entre las medias.

## **2. Prueba F de Fisher (ANOVA)**

> La prueba F de Fisher se utiliza para comparar las varianzas de tres o más grupos. En este ejemplo, realizamos un análisis de varianza (ANOVA) para comparar tres grupos.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Datos de ejemplo (tres grupos)
grupo_1 = [23, 21, 19, 24, 30, 25, 22, 24, 23, 27]
grupo_2 = [30, 29, 31, 35, 38, 40, 33, 30, 32, 35]
grupo_3 = [10, 12, 14, 15, 11, 13, 16, 14, 17, 18]

# Realizamos la prueba F (ANOVA de un factor)
f_stat, p_value = stats.f_oneway(grupo_1, grupo_2, grupo_3)

# Imprimir resultados
print("Estadístico F:", f_stat)
print("Valor p:", p_value)

# Interpretación de los resultados
alpha = 0.05  # Nivel de significancia
if p_value < alpha:
    print("Al menos uno de los grupos es significativamente diferente.")
else:
    print("No hay diferencias significativas entre los grupos.")

# Gráficos
plt.figure(figsize=(12, 6))

# Boxplot para visualizar las distribuciones de los tres grupos
plt.subplot(1, 2, 1)
plt.boxplot([grupo_1, grupo_2, grupo_3], labels=["Grupo 1", "Grupo 2", "Grupo 3"], patch_artist=True, boxprops=dict(facecolor="lightgreen", color="green"))
plt.title("Distribución de los Grupos")
plt.ylabel("Valor")
plt.grid(True)

# Gráfico de barras con las medias de los tres grupos
plt.subplot(1, 2, 2)
group_means = [np.mean(grupo_1), np.mean(grupo_2), np.mean(grupo_3)]
plt.bar(["Grupo 1", "Grupo 2", "Grupo 3"], group_means, color=['green', 'orange', 'blue'])
plt.title("Media de los Grupos")
plt.ylabel("Media")
plt.grid(True)

plt.tight_layout()
plt.show()
````

* Comparamos tres grupos diferentes para ver si sus medias son significativamente distintas.
* Al igual que en la prueba t, el valor p nos ayuda a determinar si podemos rechazar la hipótesis nula de que todas las medias son iguales.
* **Boxplot:** Muestra las distribuciones de los tres grupos, lo que facilita la comparación visual de las variabilidades y diferencias entre ellos.
* **Gráfico de Barras:** Representa las medias de los tres grupos, ayudando a observar si hay diferencias evidentes entre ellas.

### 1.1.1.Teorema 2.3 (Completo, de la página 32 de la tesis)

Sean $0 < h \leq \tau$ y $(m-1)\tau \leq t \leq m\tau - h$ para $m \geq 1$. Entonces, escribiendo $F(s) = \gamma \phi'(s) + \beta \phi(s)$, la solución $x(t)$ de (2.5), con condición inicial (2.2), satisface
$$x(t + h) = e^{\alpha h} \, x(t) + \sum_{k=1}^{m-1} \sum_{r=1}^{k} \gamma^{k-r} (\beta + \alpha \gamma)^r \frac{h^r}{r!} \binom{k-1}{r-1} x(t - k\tau)$$
$$+ \int_{t - m\tau}^{t - m\tau + h} \left[ \gamma^{m-1} + \sum_{k=1}^{m-1} \sum_{r=1}^{k} \gamma^{m-1-r} (\beta + \alpha \gamma)^r \frac{(t - m\tau + h - s)^r}{r!} \right] e^{\alpha (t - m\tau + h - s)} F(s) \, ds$$

* Esta expresión deriva de una construcción directa de esquemas numéricos exactos para DDEs lineales de tipo neutral de primer orden: $x'(t) - \gamma x'(t - \tau) = \alpha x(t) + \beta x(t - \tau)$, con $x(t) = \phi(t)$ para $-\tau \leq t \leq 0$.
* La demostración involucra representaciones vectoriales y matrices exponenciales, generalizando resultados previos para casos sin neutralidad ($\gamma = 0$).
* Es "exacta" porque reproduce la solución continua en intervalos discretos, preservando propiedades dinámicas como estabilidad dependiente del retardo.

🧪 **INTERPRETACIÓN:**

Proporciona una expresión continua exacta para la solución de una DDE neutral lineal en un paso futuro $t + h$. Incluye un término exponencial para la dinámica local, una doble suma para acumular efectos de retardos pasados (ponderados por coeficientes binomiales y factoriales), y una integral que incorpora la condición inicial. Esto asegura precisión sin errores de aproximación, ideal para validar estabilidad en sistemas cuánticos donde retardos inducen oscilaciones.

### 1.1.2. Corolario 2.1 (Completo, de la página 33 de la tesis)

Sea $h > 0$, tal que $Nh = \tau$ para algún entero $N \geq 1$. Escribiendo $t_n \equiv nh$ y $x_n \equiv x(t_n)$ para $n \geq -N$, la solución numérica exacta de (2.5) con la condición inicial (2.2), para $-N \leq n \leq 0$ está dada por $x_n = \phi(t_n)$, y para $(m-1)\tau \leq nh < m\tau$ y $m \geq 1$ por
$$x_{n+1} = e^{\alpha h} x_n + \sum_{k=1}^{m-1} \sum_{r=1}^{k} \gamma^{k-r} (\beta + \alpha \gamma)^r \frac{h^r}{r!} \binom{k-1}{r-1} x_{n - kN}$$
$$+ \int_{t_n - m\tau}^{t_n - m\tau + h} \left[ \gamma^{m-1} + \sum_{k=1}^{m-1} \sum_{r=1}^{k} \gamma^{m-1-r} (\beta + \alpha \gamma)^r \frac{(t_n - m\tau + h - s)^r}{r!} \right] e^{\alpha (t_n - m\tau + h - s)} F(s) \, ds$$

* Este corolario discretiza el Teorema 2.3 en una malla temporal donde el retardo $\tau$ es múltiplo del paso $h$, convirtiéndolo en un esquema numérico recursivo.
* Para implementación, la integral se evalúa numéricamente (e.j., con cuadraturas), pero el esquema es exacto si se computa perfectamente, evitando errores de discretización.
* Es el primer esquema exacto reportado para DDEs neutrales, extendiendo trabajos previos [24] para casos retardados.

🧪 **INTERPRETACIÓN:**

Versión discreta del teorema, adaptada a una malla temporal (con retardo como múltiplo de $h$). Calcula valores numéricos $x_{n+1}$ recursivamente, reproduciendo exactamente la solución continua en puntos de la malla. En simulación de qubits, esto permite comparar métricas como fidelidad (usando ANOVA) entre modelos con/sin retardo, validando si efectos de memoria afectan significativamente la decoherencia.

### 1.1.3. Ejemplo de Aplicación: Validación Cuantitativa de Simulaciones de Qubits con Retardo

> Modelaremos la observable de un qubit (por ejemplo, el valor esperado del operador $Z$ de Pauli, $\langle \sigma_z \rangle$) utilizando una DDE lineal neutra similar a la ecuación (2.5) de la tesis: $ x' (t) - \gamma x'(t - \tau) = \alpha x(t) + \beta x(t - \tau) $, donde $x(t)$ aproxima la dinámica de decoherencia del qubit influenciada por una retroalimentación retardada (por ejemplo, debido a efectos de memoria no markovianos en un circuito cuántico). Implementaremos el Corolario 2.1 para calcular la solución numérica exacta y compararla con una simulación markoviana (sin retraso) utilizando QuTiP. El ANOVA validará si el retraso afecta significativamente al comportamiento del qubit, en consonancia con el tema de la validación cuantitativa.

````python
import numpy as np
import qutip as qt
from scipy.integrate import quad
from scipy.special import comb
from math import factorial
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd
import matplotlib.pyplot as plt

# Parámetros del modelo DDE
alpha = -1.0
beta = 0.5
gamma = 0.75
tau = 1.0
h = tau / 10
N = int(tau / h)
t_max = 5 * tau
n_steps = int(t_max / h)

# Función inicial phi(t)
def phi(t):
    return (t + 1)**2

def phi_prime(t):
    return 2 * (t + 1)

def F(s):
    return gamma * phi_prime(s) + beta * phi(s)

# Esquema exacto DDE
def exact_dde_step(x_prev, n, m):
    term1 = np.exp(alpha * h) * x_prev[n]

    term2 = 0
    for k in range(1, m):
        for r in range(1, k+1):
            binom = comb(k-1, r-1, exact=True)
            term2 += (gamma**(k-r)) * ((beta + alpha * gamma)**r) * (h**r / factorial(r)) * binom * x_prev[n - k*N]

    def integrand(s):
        inner_sum = gamma**(m-1)
        for k in range(1, m):
            for r in range(1, k+1):
                inner_sum += (gamma**(m-1-r)) * ((beta + alpha * gamma)**r) * (((n*h - m*tau + h - s)**r) / factorial(r))
        return inner_sum * np.exp(alpha * (n*h - m*tau + h - s)) * F(s)

    lower = n*h - m*tau
    upper = lower + h
    term3, _ = quad(integrand, lower, upper)

    return term1 + term2 + term3

# Simulación DDE
t = np.linspace(-tau, t_max, n_steps + N + 1)
x = np.zeros(len(t))
for i in range(N+1):
    x[i] = phi(t[i])

for n in range(0, n_steps):
    m = int(np.floor((n*h)/tau) + 1)
    x[n + N + 1] = exact_dde_step(x[:n + N + 1], n, m)

# Simulación cuántica (QuTiP) sin retardo
H = qt.sigmax()
psi0 = qt.basis(2, 1)
c_ops = [np.sqrt(-alpha) * qt.sigmaz()]
t_qutip = t[t >= 0]
result_no_delay = qt.mesolve(H, psi0, t_qutip, c_ops=c_ops, e_ops=[qt.sigmaz()])
sz_no_delay = result_no_delay.expect[0]

# 🔍 Ajustar longitudes
x_delay = x[N + 1:N + 1 + len(sz_no_delay)]
min_len = min(len(x_delay), len(sz_no_delay), len(t_qutip))

x_delay = x_delay[:min_len]
sz_no_delay = sz_no_delay[:min_len]
t_qutip = t_qutip[:min_len]

# Preparar DataFrame para ANOVA
data = pd.DataFrame({
    'Tiempo': np.tile(t_qutip, 2),
    'Valor': np.concatenate([x_delay, sz_no_delay]),
    'Grupo': ['Con Retardo'] * min_len + ['Sin Retardo'] * min_len
})

# ANOVA
model = ols('Valor ~ C(Grupo)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Visualización
plt.figure(figsize=(10, 6))
plt.plot(t_qutip, x_delay, label='Con Retardo (Esquema Exacto DDE)', color='blue')
plt.plot(t_qutip, sz_no_delay, label='Sin Retardo (QuTiP)', color='red', linestyle='--')
plt.xlabel('Tiempo')
plt.ylabel('Expectativa <σ_z>')
plt.title('Simulación de Qubit: Con vs. Sin Retardo')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Mostrar tabla ANOVA
print("Tabla ANOVA:\n", anova_table)
````

**1. Gráfico:**
* El código genera un gráfico comparando dos curvas:
    * Azul (Con Retardo): La solución numérica exacta del DDE usando el Corolario 2.1, representando la expectativa $\langle \sigma_z \rangle$ del qubit con memoria no-Markoviana.
    * Rojo (Sin Retardo): La solución Markoviana (QuTiP) sin retardo, que modela decoherencia estándar.

**2. Tabla ANOVA:**

* `sum_sq:` Suma de cuadrados entre grupos (Con Retardo vs. Sin Retardo) y residual.

🧠 **INTERPRETACIÓN:**

* El esquema exacto del Corolario 2.1 (basado en Teorema 2.3) asegura que la simulación del qubit con retardo reproduce fielmente la dinámica teórica, sin errores de discretización. Esto es crucial en ingeniería cuántica, donde errores numéricos pueden simular decoherencia artificial.
* ANOVA confirma estadísticamente si el retardo introduce diferencias significativas en $\langle \sigma_z \rangle$, lo que valida experimentalmente el impacto de efectos no-Markovianos en qubits (e.g., en circuitos cuánticos con feedback retardado).
* Esto alinea con el tema de la exposición: los métodos estadísticos cuantitativos (como ANOVA) son esenciales para validar simulaciones numéricas en ciencia e ingeniería, especialmente en aplicaciones emergentes como la computación cuántica.

# 2.  Interpolación con Matriz de Vandermonde

<a id="2.1"></a>
## 2.1. Introducción

La interpolación es una técnica fundamental en el análisis cuantitativo de datos experimentales y simulaciones. En particular, la **matriz de Vandermonde** permite construir un polinomio que pasa exactamente por un conjunto de puntos conocidos. Esta herramienta es útil en ciencia e ingeniería para validar modelos, simular resultados intermedios y analizar el comportamiento de sistemas físicos.

Desde una perspectiva epistemológica, la interpolación apoya la **validez interna** de los modelos, al garantizar que las predicciones se ajustan a los datos empíricos observados, dentro de ciertos márgenes de error y supuestos.

<a id="2.2"></a>
## 2.2. Definición Matemática

Dado un conjunto de $( n )$ puntos $(x_1, y_1), (x_2, y_2), \dots, (x_n, y_n)$ , la interpolación polinomial busca encontrar un polinomio $( P(x) )$ de grado $( n-1 )$ tal que:

<div align="center">

$P(x_i) = y_i$ para $i = 1, 2, \dots, n$

</div>

Este polinomio puede representarse en la forma:

<div align="center">

$P(x) = a_0 + a_1x + a_2x^2 + \dots + a_{n-1}x^{n-1}$

</div>

Para encontrar los coeficientes $( a_0, a_1, \dots, a_{n-1} )$, se forma un sistema lineal:

<div align="center">

$\begin{bmatrix}
1 & x_1 & x_1^2 & \dots & x_1^{n-1} \\
1 & x_2 & x_2^2 & \dots & x_2^{n-1} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & x_n & x_n^2 & \dots & x_n^{n-1}
\end{bmatrix}
\begin{bmatrix}
a_0 \\
a_1 \\
\vdots \\
a_{n-1}
\end{bmatrix}
=
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_n
\end{bmatrix}$


</div>

La matriz a la izquierda es la **matriz de Vandermonde**, denotada como $( V )$. La solución del sistema $( V \cdot \mathbf{a} = \mathbf{y} )$ nos da los coeficientes del polinomio interpolador.


<a id="2.3"></a>
## 2.3. Aplicación en Análisis Cuantitativo

En un contexto de investigación cuantitativa, la interpolación con la matriz de Vandermonde permite:

- **Validar modelos**: Comparando resultados de simulación con datos experimentales interpolados.
- **Estimar variables intermedias**: Cuando se tienen pocos puntos medidos y se requiere conocer el comportamiento del sistema en puntos no observados.
- **Analizar confiabilidad**: Evaluar la sensibilidad del modelo a cambios en las entradas.
- **Explorar validez externa**: Generalizar resultados a partir de los datos conocidos.

<a id="2.4"></a>
## 2.4. Ejemplo: Dataset de Simulación de Temperatura en un Dispositivo Cuántico

Supongamos que en un experimento simulado se mide la **temperatura interna** de un componente cuántico en tres momentos del tiempo:


<div align="center">

| Tiempo (s) | Temperatura (°C) |
|------------|------------------|
| 0.0        | 22.0             |
| 1.0        | 24.5             |
| 2.0        | 29.0             |

</div>

Queremos interpolar un polinomio que permita estimar la temperatura a cualquier instante dentro del intervalo $([0, 2])$.


````python
import numpy as np
import matplotlib.pyplot as plt

# Datos experimentales
x = np.array([0.0, 1.0, 2.0])     # Tiempo
y = np.array([22.0, 24.5, 29.0])  # Temperatura

# Crear matriz de Vandermonde
V = np.vander(x, increasing=True)

# Resolver el sistema lineal
a = np.linalg.solve(V, y)

# Mostrar coeficientes del polinomio
print("Coeficientes del polinomio interpolador:")
for i, coef in enumerate(a):
    print(f"a{i} = {coef:.4f}")

# Crear función polinomial
def P(t):
    return sum(a[i] * t**i for i in range(len(a)))

# Graficar
t_vals = np.linspace(0, 2, 100)
temp_vals = P(t_vals)

plt.plot(t_vals, temp_vals, label='Polinomio Interpolador')
plt.plot(x, y, 'ro', label='Datos Originales')
plt.xlabel('Tiempo (s)')
plt.ylabel('Temperatura (°C)')
plt.title('Interpolación con Matriz de Vandermonde')
plt.legend()
plt.grid(True)
plt.show()
````

✅ **Interpretación de Resultados**

* El polinomio resultante es de grado 2 (parabólico), ajustado a los tres puntos.
* Podemos usar este modelo para estimar la temperatura en tiempos no medidos, por ejemplo:

````python
print(f"T(1.5) = {P(1.5):.2f} °C")
````

* Esto apoya la identificación de variables dependientes (temperatura) y variables independientes (tiempo).
* Permite establecer relaciones causales entre variables en un modelo cuantitativo.

<a id="2.5"></a>
## 2.5. Conexión Epistemológica

* **Confiabilidad:** La interpolación es determinista; el resultado es reproducible si los datos son los mismos.
* **Validez interna:** El modelo interpola exactamente los datos, pero su extrapolación fuera del rango puede ser poco confiable.
* **Identificación de variables:** Se clarifica la relación entre tiempo y temperatura.
* **Alcance de la investigación:** El modelo interpolado solo es válido en el intervalo de datos $[0,2]$ segundos, fuera de este puede no representar el fenómeno real.

\subsection{Objetivo}
El objetivo de este ejercicio es interpolar un polinomio de segundo grado que pase por tres puntos dados utilizando la matriz de Vandermonde.

\subsection{Método}
Utilizando la matriz de Vandermonde, se construye el sistema de ecuaciones para los tres puntos dados y se resuelve el sistema lineal para obtener los coeficientes del polinomio.

\begin{lstlisting}[language=Python, caption=Interpolación con la Matriz de Vandermonde]
import numpy as np
import matplotlib.pyplot as plt

# Datos experimentales
x = np.array([0.0, 1.0, 2.0])
y = np.array([22.0, 24.5, 29.0])

# Construcción de la matriz de Vandermonde
V = np.vander(x, increasing=True)

# Resolución del sistema para obtener los coeficientes del polinomio
coefficients = np.linalg.solve(V, y)

# Evaluación del polinomio sobre un rango de valores
x_plot = np.linspace(0, 2, 100)
y_plot = np.polyval(coefficients[::-1], x_plot)

# Gráfico del polinomio
plt.plot(x_plot, y_plot, label='Polinomio interpolado')
plt.scatter(x, y, color='red', label='Puntos originales')
plt.xlabel('Tiempo (s)')
plt.ylabel('Temperatura (°C)')
plt.title('Interpolación con la Matriz de Vandermonde')
plt.legend()
plt.show()
\end{lstlisting}
